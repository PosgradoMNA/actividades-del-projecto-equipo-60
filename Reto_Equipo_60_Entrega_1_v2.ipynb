```python
#Creamos el dataframe "dfnum" solo con las columnas numéricas del conjunto de datos original.
dfnum = dfAguasSubterraneas[ColumnasNumericas]

#Creamos una lista para almacenar todos los valores no numéricos que encontramos en las columnas del nuevo dataframe
list_nn = list()
for col in range(len(dfnum)):
  for fil in dfnum.iloc[col]:
     tem=str(fil)
     if(tem.isnumeric()==False):
        #Asignamos el valor no numérico encontrado a la lista.
        list_nn.append(fil)  

#Convertimos todos los valores de la lista a 'str'
list_nn_final=list(map(str,list_nn))
#Ordenamos la lista en forma descendente para obtener aquellos valores con caracteres no numéricos al comienzo.
list_nn_final.sort(reverse=True)
print("*********************************************************************************************************")
print("Las siguientes son ocurrencias por valor no numérico encontradas en columnas que deberían ser numéricas:")
print("*********************************************************************************************************")
pd.DataFrame(list_nn_final).sort_values([0]).value_counts().head(15)
```

*   <20 = 20

*   <0.2 = 0.2
*   <0.025 = 0.025
*   <0.02 = 0.02
*   <0.01 = 0.01
*   <0.005 = 0.005
*   <0.003 = 0.003
*   <0.0015 = 0.0015
*   <0.0005 = 0.0005


Una vez ajustados estos valores procederemos a cambiar el tipo de dato para las columnas numéricas.

✍ A continuación implementaremos el modelo **KMeans** que nos permitirá agrupar los puntos de agua indicados en el conjunto de datos en **K** clústeres basándose en una característica que los agrupe, para este ejercicio la característica será la coordenada (longitud y latitud) de cada punto.

Una vez obtenidos los clústeres, pasaremos a graficar los puntos agrupados mostrando el color indicado en el semáforo (verde, amarillo o rojo) que nos indicará el grado de contaminación del agua para el punto relacionado, de tal forma que podamos concluir si existe una relación entre la calidad del agua y la ubicación geográfica de los puntos indicados en el conjunto de datos. El significado de los colores en el semáforo se muestra a continuación:

🔴 Contaminada.
🟡 Medianamente potable.
🟢 Potable.

---

```python
#Iniciamos creando un nuevo dataframe llamado "df_coord" con las columnas que conforman la característica de agrupación para los clústeres 
#y el valor del semáforo por cada coordenada (latitud y longitud).
#Pasamos el nombre del color a inglés para que podamos contar con un valor estándar y validar si podemos usarlos más adelante en la graficación.
df_coord = dfAguasSubterraneas[['LATITUD','LONGITUD','SEMAFORO']]
df_coord['SEMAFORO']= df_coord['SEMAFORO'].str.replace('Verde','green')
df_coord['SEMAFORO']= df_coord['SEMAFORO'].str.replace('Amarillo','yellow')
df_coord['SEMAFORO']= df_coord['SEMAFORO'].str.replace('Rojo','red')
```

```python
#Ejecutamos el visualizador KElbow con el fin de encontrar el número óptimo de clústeres de acuerdo a la característica y tamaño de muestras.
from yellowbrick.cluster import KElbowVisualizer
from sklearn.cluster import KMeans
from sklearn.datasets import make_blobs

blob_centers = list(zip(df_coord.LONGITUD, df_coord.LATITUD))

X, y = make_blobs(n_samples=1068, centers=blob_centers, cluster_std=0.20,
                  random_state=42)

model = KElbowVisualizer(KMeans(), k=(2,25))
model.fit(X)
model.show()
```

---


🔎 Observamos que el visualizador KElbow nos indica que el número óptimo de clústeres (**K**) basado en la característica común (coordenadas) y el número de muestras de nuestro conjunto de datos (1.068) es **6**.
Siguiendo la recomendación de KElbow procederemos a ejecutar nuestro modelo KMeans haciendo uso de los 6 clústeres indicados.

---

```python
#Capturamos en la variable K el número óptimo de clústeres
K = model.elbow_value_
#Ejecutamos el KMeans con los K clústeres y capturamos los resultados en la variable "KMeansOptimo".
KMeansOptimo = KMeans(n_clusters=K, random_state=42)
#Realizamos la predicción de nuestra y basados en el KMeans óptimo y los atributos de entrada en X.
y_pred = KMeansOptimo.fit_predict(X)
#Imprimirmos el valor de la predicción.
print("******************************************")
print("Los valores de predicción para KMeans son:")
print("******************************************")
print(y_pred)
print("")
#Capturamos en la variable "Clusters" e imprimimos los centroides para el clústeres encontrados en la ejecución del KMeans óptimo.
Clusters=KMeansOptimo.cluster_centers_
print("***********************************************************")
print("Los centroides se encuentran en las siguientes coordenadas:")
print("***********************************************************")
print(Clusters)
```
